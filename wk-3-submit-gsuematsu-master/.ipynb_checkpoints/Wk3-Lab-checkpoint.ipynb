### `scipy.optimize.least_squares`
Following the document of [`scipy.optimize.least_squares`](https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.least_squares.html)

```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
from scipy.special import binom
from scipy.optimize import brentq

from scipy.optimize import least_squares
```

```python
def fun_rosenbrock(x):
    return np.array([10 * (x[1] - x[0]**2), (1 - x[0])])

import matplotlib.image as mplimg
im_path = '1920px-Rosenbrock_function.png'
fig3 = mplimg.imread(im_path)
plt.figure(figsize =(10,10));
plt.imshow(fig3);
```

```python
x0_rosenbrock = np.array([2, 2])
res_1 = least_squares(fun_rosenbrock, x0_rosenbrock)
res_1.x
```

```python
res_2 = least_squares(fun_rosenbrock, x0_rosenbrock, bounds=[(0,2), (2,3)])
res_2.x
```

### `sklearn.model_selection.LeaveOneOut`

```python
from sklearn.model_selection import LeaveOneOut
X = np.array([[1, 2], [3, 4]])
y = np.array([1, 2])
loo = LeaveOneOut()

print(loo)

for train_index, test_index in loo.split(X):
    print("TRAIN:", train_index, "TEST:", test_index)
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    # Actually you need your validation function here 
    print(X_train, X_test, y_train, y_test)
```
