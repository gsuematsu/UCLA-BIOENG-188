### (1) Reproduce all four panels from Figure 1 shown in Perelson et al. using eq. 6

Since we are not given any empirical data, use the fitted parameters and the appropriate constants to generate the curves shown on these plots. Remember to account for the time delays for each patient.

#### (i) Implement eq. 6 from the paper as a function.

```python
### Answer
# Define function outlined as equation 6
def plasmaVirionConcEq6(time, params):
    # Params should be given as array [V_0, c, delta]
    V_0 = params[0]
    c = params[1]
    delta = params[2]
    
    term1V = V_0 * np.exp(-c * time)
    cDivCMinusDel = c / (c - delta)
    term2V = V_0 * cDivCMinusDel**2 * (np.exp(-delta * time) - np.exp(-c * time))
    term3V = V_0 * cDivCMinusDel * -delta * time * np.exp(-c * time)
    V = term1V + term2V + term3V
    
    return V

# Create arrays for storing parameters for each patient from Table 1 [V_0, c, delta]
patient104Params = [52e3, 3.68, 0.50]
patient105Params = [643e3, 2.06, 0.53]
patient107Params = [77e3, 3.09, 0.50]

# Check that the time = 0 value is the initial value of the equation
assert(plasmaVirionConcEq6(0, patient104Params) == patient104Params[0])
assert(plasmaVirionConcEq6(0, patient105Params) == patient105Params[0])
assert(plasmaVirionConcEq6(0, patient107Params) == patient107Params[0])
```

Biol. 192:81-98) to be $3.43\times10^{-8}$ mL/(virion•day). Note that you will need to solve ordinary differential equations for $T^*$, $V_I$, and $V_{NI}$ to reproduce the data in Figure 1.

### (3) Reimplement the model relaxing the assumption about T remaining constant by numerically solving the system of ODEs.

#### i. Define a function that takes each parameter as a vector and outputs the time derivative of each species.

```python
# Answer
from scipy.integrate import odeint

def afterDrugODE(y, time, k, delta, c, N, Ttot):
    # params = [k, delta, c, N, Ttot]
#     k = params[0]
#     delta = params[1]
#     c = params[2]
#     N = params[3]
#     Ttot = params[4]
    
    T_Star, V_I, V_NI = y

    dTStardt = k * V_I * (Ttot - T_Star) - delta * T_Star
    dV_Idt = -c * V_I
    dV_NIdt = N * delta * T_Star - c * V_NI
    
    return np.array([dTStardt, dV_Idt, dV_NIdt])
    
```

#### ii. Use odeint to solve this system of equations numerically for patient 105.

Hint: For ease later, it may be helpful to make a function here that handles everything but takes in gnu.

```python
# Answer
k = 3.43e-8
# Create arrays for storing parameters for each patient from Table 1 [V_0, c, delta]
patient104Params = [52e3, 3.68, 0.50]
patient105Params = [643e3, 2.06, 0.53]
patient107Params = [77e3, 3.09, 0.50]

patient105Ttot = 11e3
patient105delta = patient105Params[2]
patient105c = patient105Params[1]

# Use steady state eqns to calculate the N, TStar and T inital conditions
# Following eqns were solved analytically
# k*V*T - delta*TStar = 0
# N*delta*TStar - c*V = 0
# Ttot - TStar - T = 0
# T is defined as V_NI
def steadyState(k, V_0, delta, c, Ttot):
    TStarSteady = (Ttot * k * V_0)/(k * V_0 + delta)
    V_NISteady = Ttot - TStarSteady
    NSteady = (c * V_0)/(delta * TStarSteady)
    
    return np.array([TStarSteady, V_NISteady, NSteady])

patient105SteadyStateValues = steadyState(k, patient105Params[0], patient105delta, patient105c, patient105Ttot)
print('Steadystate outputs for patient 105:', patient105SteadyStateValues)
# Store ICs in array [T_Star, V_I, V_NI]
patient105TStarIC = patient105SteadyStateValues[0]
patient105V_IIC = patient105Params[0]
patient105V_NIIC = 0
patient105ICs = np.array([patient105TStarIC, patient105V_IIC, patient105V_NIIC])

# Parameters are given as [k, delta, c, N, Ttot]
patient105ODEParams = (k, patient105delta, patient105c, patient105SteadyStateValues[2], patient105Ttot)
patient105solution = odeint(afterDrugODE, patient105ICs, t, args = patient105ODEParams)
```

#### iii. Reproduce the subpanels of Fig. 1 with this updated model. 

```python
# Answer
# Plot data for patient 105
patient105ODESolverVtot = 2 * (patient105solution[:, 1] + patient105solution[:, 2])
plt.figure(figsize = (20, 5))
plt.subplot(132)
plt.semilogy(t - patient105TimeDelay, patient105ODESolverVtot)
plt.xlim(0, 8)
plt.ylim(1e4, 1e7)
plt.xlabel('Days')
plt.ylabel('RNA copies/mL')
plt.title('Patient 105')

plt.show()
```

### (3) Repeat the work that the authors describe in item 12 of their “References and Notes”.

#### (i) Namely, vary the effectiveness factor of the drug (η=1.0, 0.99, 0.95, and 0.90) and simulate the viral load that would result from the modified differential equations using the parameters c=3.0 days$^{-1}$ and δ=0.5 days$^{-1}$ as described in item 12 and the values of T0, V0, and the time delay provided for patient 105.

```python
# Answer
etas = np.array([1.0, 0.99, 0.95, 0.90])
patient105VvaryingEffFac = np.zeros((len(t), len(etas)))

# Store new values according to item 12
patient105Params_12 = np.array([patient105Params[0], 3.0, 0.50])

# Use steady state function to calculate new steady state values with new parameters
patient105SteadyStateValues_12 = steadyState(k, patient105Params_12[0], patient105Params_12[2], patient105Params_12[1], patient105Ttot)

# Store ICs in array [T_Star, V_I, V_NI]
patient105TStarIC_12 = patient105SteadyStateValues_12[0]
patient105V_IIC_12 = patient105Params_12[0]
patient105V_NIIC_12 = 0
patient105ICs_12 = np.array([patient105TStarIC_12, patient105V_IIC_12, patient105V_NIIC_12])

# Define new function that takes effectiveness factor as a paramter
def afterDrugODE_eta(y, time, k, delta, c, N, Ttot, eta):
    # params = [k, delta, c, N, Ttot]
#     k = params[0]
#     delta = params[1]
#     c = params[2]
#     N = params[3]
#     Ttot = params[4]
    
    T_Star, V_I, V_NI = y

    dT_Stardt = k * V_I * (Ttot - T_Star) - delta * T_Star
    dV_Idt = -c * V_I + ((1 - eta) * N * delta * T_Star)
    dV_NIdt = N * delta * T_Star * eta  - c * V_NI
    
    return np.array([dT_Stardt, dV_Idt, dV_NIdt])

for i in range(len(etas)):
    # Create parameters for new dynamic model
    patient105ODEParams_12 = (k, patient105Params_12[2], patient105Params_12[1], patient105SteadyStateValues_12[2], patient105Ttot, etas[i])
    
    # Solve ODE using new parameters
    patient105etaSolution = odeint(afterDrugODE_eta, patient105ICs_12, t, patient105ODEParams_12)
    
    # Store new Vs in matirx
    patient105VvaryingEffFac[:, i] = 2 * (patient105etaSolution[:, 1] + patient105etaSolution[:, 2])
    
# Plot Vs with varying etas
plt.figure(figsize = (10, 10))
labels = ['eta = 1.0', 'eta = 0.99', 'eta = 0.95', 'eta = 0.90']
for i in range(len(etas)):
    plt.semilogy(t - patient105TimeDelay, patient105VvaryingEffFac[:, i], label = labels[i])

plt.legend()
plt.xlim(0,8)
plt.ylim(1e4, 1e7)
plt.xlabel('Days')
plt.ylabel('RNA copies/mL')
plt.title('Patient 105')
plt.show()
```

#### (ii) Then, use the function `scipy.optimize.leastsq` to fit this data to the equation for V(t) given in the paper and find the estimates of c and δ that result. Fit only the portion of the curve after the pharmacokinetic delay. Compare these estimates to the actual values for c and δ and discuss the consequences of an imperfect drug on clearance time estimates.

```python
# Answer
from scipy.optimize import leastsq 

# Start parameters for least squares [c, delta]
startParams = np.array([3.0, 0.5])

# New time array for patient 105
patient105t = np.linspace(patient105TimeDelay, 8, 20)

# Matrices to store betas for time delay and no time delay
patient105BetaNoTimeDelay = np.zeros((len(etas), len(startParams)))
patient105BetaTimeDelay = np.zeros((len(etas), len(startParams)))

# Define function to calculate residuals
def residuals(varyParams, V_0, residualTime, measurements):
    residualParams = np.array([V_0, varyParams[0], varyParams[1]])
    pred = 2 * plasmaVirionConcEq6(residualTime, residualParams)
    return (pred - measurements)

def ls(residualsLS, paramsLS, argsLS):
    LSqVals = leastsq(residualsLS, paramsLS, args=argsLS)
    return LSqVals[0]    
    
patient105Params = [643e3, 2.06, 0.53]

for i in range(len(etas)):
    # LS with no time delay
    patient105lsNoTD = ls(residuals, startParams, argsLS = (patient105Params[0], t, np.transpose(patient105VvaryingEffFac[:, i])))
    patient105BetaNoTimeDelay[i] = patient105lsNoTD
    
    # LS with time delay
    patient105lsTD = ls(residuals, startParams, argsLS = (patient105Params[0], patient105t, np.transpose(patient105VvaryingEffFac[:, i])))
    patient105BetaTimeDelay[i] = patient105lsTD

# Print betas found for time delay and no time delay for patient 105       
import pandas as pd
etasCol = np.array([[etas[0]], [etas[1]], [etas[2]], [etas[3]]])
print("Least Squares Solution for Patient 105 with no Time Delay")
header = ('eta', 'c', 'delta')
dataNoTD = np.concatenate((etasCol, patient105BetaNoTimeDelay), axis = 1)
tableNoTD = pd.DataFrame(dataNoTD, columns = header)
print(tableNoTD)  

print("Least Squares Solution for Patient 105 with Time Delay")
dataTD = np.concatenate((etasCol, patient105BetaTimeDelay), axis = 1)
tableTD = pd.DataFrame(dataTD, columns = header)
print(tableTD)                 
```

#### When observing the results of the least squares solution when the time delay is not implemented, the c and delta values are similar to the c and delta values found in item 12 of Perelson et al. When the time delay is accounted for the least squares output gives almost the exact c = 2.06 and delta = 0.53 presented in Table 1 of Perelson et al. 
#### As the drug efficiency (eta) decreaes from 1.0 both situations (time delay and no time delay) displays and increase in the clearance (c) and the loss rate of virion producing cells (delta) decreases. The V(t) in equation 6 shows an exponential decay of V_0exp(-ct), which indicates that with imperfect drugs the concentration of virion will no decay as quickly as c decreases. The remainder of Eq 6 possesses dependency on delta and c, and cannot make an immediate conjecture on the effect of an increase in delta when c decreases.

#### (iii) Examine the residual plots from your fits to determine whether or not there is a trend in the error associated with fitting these new curves.

```python
# Answer
residualsWithVaryingEta = np.zeros((len(etas), len(t)))

for i in range(len(etas)):
    residualsWithVaryingEta[i] = residuals(patient105BetaTimeDelay[i], patient105Params[0], t, np.transpose(patient105VvaryingEffFac[:, i]))

# Plot residuals for each error value
plt.figure()
plt.plot(t, residualsWithVaryingEta[0], label = 'eta = 1.0')
plt.plot(t, residualsWithVaryingEta[1], label = 'eta = 0.99')
plt.plot(t, residualsWithVaryingEta[2], label = 'eta = 0.95')
plt.plot(t, residualsWithVaryingEta[3], label = 'eta = 0.90')
plt.legend()
plt.show()
```

#### The residuals plot is a way to estimate the fitting error of our model. The plot shows that as the model increases with time the error dramatically increases, but after day 2 the fitting error approaches zero. Additionally, the residual error is higher for higher etas. This is because with the eta = 1.0 situation, we are assuming a perfect drug, which is not the case. As the eta becomes smaller and closer to a more realistic situation the residual error decreases. In order to find the eta in effect, least squares could be run again where we allow c delta and eta to vary and find the least squares  solution for all three.
