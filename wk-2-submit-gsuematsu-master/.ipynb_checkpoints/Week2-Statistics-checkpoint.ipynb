## (1) Estimation of a sample mean from a normally distributed variable.

Let us assume that a true distribution of a process is described by the normal distribution with $\mu=5$ and $\sigma=1$. You have a measurement technique that allows you to sample n points from this distribution. In Python we can use a random number generator whose numbers will be chosen from the desired normal distribution by using the function `np.random.normal`. Sample from this normal distribution from n=1 to 50 (i.e. n=1:50). Create a plot for the standard deviation of the calculated mean from each n when you repeat the sampling 1000 times each. (i.e. You will repeat your n observations 1000 times and will calculate the sample mean for each of the 1000 trials).

```python
# 
mean = 5
stddev = 1
standDevs = np.zeros(50)
means = np.zeros((50, 1000))
for i in range(0, 50):
    n = i + 1
    
    for j in range(0, 1000):
        X = np.random.normal(mean, stddev, n)
        means[i, j] = np.mean(X)
    
    standDevs[i] = np.std(means[i])
    
print(standDevs)

n = np.linspace(1, 50, 50)
plt.scatter(n, standDevs)
plt.xlabel("n")
plt.ylabel("standard deviation")
plt.show()
```

#### 1a. Plot the standard deviation of the sample mean versus n. Add a second line which is 1/sqrt(n). Describe what this tells you about the relationship between n and your power to estimate the underlying mean.

```python
# Answer to 1a here
plt.scatter(n, standDevs)
plt.plot(n, 1/np.sqrt(n), c = 'r')
plt.legend(["standDevs", "1/sqrt(n)"])
plt.xlabel("n")
plt.show()
```

The plot explicitly tells me that the larger my n is that smaller my standard deviation becomes. This means that estimating the underlying mean is more accurate when estimating using the raw data.

```python
# Answer to 1b here

plt.boxplot(np.ndarray.transpose(means))
plt.show()
```

#### 1c. For n=3, plot the histogram of the mean for the 1000 trials. Use the Kolmogorov-Smirnov test to see if this sample distribution is normal. (Hint: you may need to translate this to the standard normal distribution.) Report the sample mean and sample standard deviation, the p-value from the test, and whether you would reject the null hypothesis.

```python
# Answer to 1c here
n = 3
i = n - 1
plt.hist(means[i])
plt.ylabel("# of trials")
plt.xlabel("mean")
plt.show()

x = zscore(means[i])
ks = kstest(x, "norm")
# print(ks)

# n = 3 [mean of means, standard deviation of means, p value]
n3 = [np.mean(means[i]), standDevs[i], ks[1]]
print("sample mean =", n3[0])
print("sample standard deviation =", n3[1])
print("p-value =", n3[2])
```

## (3) Differential expression . In this problem you will use the two-sample t-test to explore what differential hypothesis testing looks like in known standards and how multiple hypothesis correction effects the number of false positives and negatives from these tests.
- Distribution 1, normal with mu=1, sigma=1
- Distribution 2, normal with mu=3, sigma=1

#### 3a. False Negative: Using n=3, perform 100 comparisons of distribution 1 versus distribution 2 with an alpha=0.05. Anytime you fail to reject the hypothesis it is a false negative. Why is this a false negative? Report the number of false negatives from your 100 tests.

Hint: It'd be helpful to define a function that does this for you at this point.

```python
# Answer to 3a here.
mu1 = 1 
mu2 = 3
n = 3
std = 1

alpha = 0.05
trials = 100

def tTestFN(alpha, numTrials, n, std):
    falseNegCounter = 0;
    
    for i in range (0, numTrials):
        distr1 = np.random.normal(mu1, std, n)
        distr2 = np.random.normal(mu2, std, n)
        tTest = ttest_ind(distr1, distr2) 
        pVal = tTest[1]
        
        if pVal > alpha:
            falseNegCounter += 1
            
    return falseNegCounter
            
print("false negatives =", tTestFN(alpha, trials, n, std))
        
```

#### This is considered a false negative because it the failure to reject the null hypothesis suggests the distributions are the same, when they are different.

#### 3b. False Positives: Using n=3, perform 100 comparisons of distribution 1 versus distribution 1 with an alpha=0.05. Anytime you reject the hypothesis this is a false positive. Why is this a false positive? Report the number of false positives from your 100 tests.

```python
# Answer to 3b here.
mu1 = 1 
mu2 = 3
n = 3
std = 1

alpha = 0.05
trials = 100

def tTestFP(alpha, numTrials, n, std):
    falsePosCounter = 0
    
    for i in range (0, numTrials):
        distr1 = np.random.normal(mu1, std, n)
        distr2 = np.random.normal(mu1, std, n)
        tTest = ttest_ind(distr1, distr2) 
        pVal = tTest[1]
        
        if pVal < alpha:
            falsePosCounter += 1
    return falsePosCounter

print("false positives =", tTestFP(alpha, trials, n, std))
```

#### This is considered a false positive because the rejection of the null suggests the distributions are different when they are actually from the same distribution.

#### 3c. Repeat 3b but 1000 times. What is the number of false positives? Predict the number of false positives you would get if you compared samples from the same distribution 10,000 times and explain why.

```python
# Answer to 3c
trials = 1000
n = 3
alpha = 0.05
std = 1

print(tTestFP(alpha, trials, n, std))
```

#### I predict that as I take more sample percentage of false positives will remain the same. Therefore, I predict if I sample 10000 times the number of false positives would be close to 500.

#### 3d. Now sweep n from 3 to 30 and report the number of false positives and false negatives for each n when you run 100 comparisons. (Provide this in a table format). Please explain the trend you see and interpret its meaning.

```python
# Answer to 3d
# data contains [n, fp, fn]
data = np.zeros((30-2, 3))

alpha = 0.05
trials = 100
std = 1


for i in range (0, len(data[:, 0])):
    n = i + 3
    data[i, 0] = n
    
    data[i, 1] = tTestFP(alpha, trials, n, std)
    data[i, 2] = tTestFN(alpha, trials, n, std)
    
import pandas as pd
header = ('n', 'false positives', 'false negatives')
table = pd.DataFrame(data, columns = header)
print(table)
    
```

#### The number of false positives does not change depending on the sample size becaue it is determined by our alpha (0.05), which suggests we will have 5% false positive results. The false negatives, in contrast, rapidly drops to 0.0 when n increases. This is because the larger n is the closer the sample mean will be to the actual mean, decreasing the standard deviation. This decrease will eliminate any overlap in the sample distributions and prevent false negatives from occurring.

#### 3f. Lastly, perform 3d for p < 0.01 instead of p < 0.05. How does this influence the rate of false positives and negatives? How might you use this when performing many tests?

```python
# Answer to 3f
# data contains [n, fp, fn]
data = np.zeros((30-2, 3))

alpha = 0.01
trials = 100
std = 1

for i in range (0, len(data[:, 0])):
    n = i + 3
    data[i, 0] = n
    
    data[i, 1] = tTestFP(alpha, trials, n, std)
    data[i, 2] = tTestFN(alpha, trials, n, std)
    
import pandas as pd
header = ('n', 'false positives', 'false negatives')
table = pd.DataFrame(data, columns = header)
print(table)
```

#### As previously stated in 3d, our alpha quantity determines our false positive rate. Because alpha was decreased to 0.01, the false positives begin to appear at a 1% rate. The false negative continue to drop to 0.0 as n increases because this is determined by the means of the two distributions.

## (5) Shaffer et al

In this excercise we're going to explore some basic concepts of statistics, and use them to build up to some more advanced ideas. To examine these ideas we're going to consider a classic of molecular biologythe [Luria-Delbrck experiment](https://en.wikipedia.org/wiki/LuriaDelbrck_experiment).

```python
repOne = np.loadtxt("expt_rep1.csv")
repTwo = np.loadtxt("expt_rep2.csv")
```

#### 5a. First, we need to build up a distribution of outcomes for what an experiment would look like if it followed the Luria-Delbruck process.
Fill in the function below keeping track of normal and mutant cells. Then, make a second function, `CVofNRuns`, that runs the experiment 3000 times. You can assume a culture size of 120000 cells, and mutation rate of 0.0001 per cell per generation. What does the distribution of outcomes look like?

```python
# Runs the simulation a bunch of times, and looks for how often the fano (cv/mean) comes out to one side
cultureSize = 120000
mutationRate = 0.0001
trials = 3000

def simLuriaDelbruck(cultureSize, mutationRate):
    nCells, nMuts = 1, 0  # Start with 1 non-resistant cell

    for _ in range(np.int(np.floor(np.log2(cultureSize)))):  # num of gens
        nCells *= 2     # Double the number of cells, simulating division
        newMuts = np.random.poisson(nCells * mutationRate)  # de novo
        nMuts = nMuts * 2 + newMuts       # Previous mutants divide and add
        nCells -= newMuts    # Non-resistant pop goes down by newMuts

    return nMuts

numCells = np.zeros(trials)

def CVofNRuns(N):
    
    for i in range(0, N):
        numCells[i] = simLuriaDelbruck(cultureSize, mutationRate)
    
    stdCells = np.std(numCells)
    meanCells = np.mean(numCells)
    cv = stdCells/meanCells
    fano = cv/meanCells
    
    plt.hist(numCells)
    plt.xlabel("trials")
    plt.ylabel("number of mutant cells")
    plt.show()
    
    return [cv, fano]

[cv, fano] = CVofNRuns(trials)

print("fano =", fano)
```

#### 5b. Compare the distribution of outcomes between the two replicates of the experiment using the 2-sample KS test. Are they consistent with one another?
Hint: Each experiment varies slightly in the amount of time it was run. The absolute values of the numbers doesn't matter, so much as the variation of them. You'll need to correct for this by dividing by the mean of the results.

```python
# Answer to 5b
repOneCorrect = repOne/np.mean(repOne)
repTwoCorrect = repTwo/np.mean(repTwo)

ks2samp = ks_2samp(repOneCorrect, repTwoCorrect)
print("p-value =", ks2samp[1])
```

#### 5c. Compare the distribution of outcomes between the experiment and model. Are our results consistent with resistance arising through a Luria-Delbruck related process?

```python
# Answer to 5c
numCellsCorrect = numCells/np.mean(numCells)

ks2sampRepOne = ks_2samp(repOneCorrect, numCellsCorrect)
ks2sampRepTwo = ks_2samp(repTwoCorrect, numCellsCorrect)

print("p-value for repOne =", ks2sampRepOne[1])
print("p-value for repTwo =", ks2sampRepTwo[1])

```

#### The 2 sample KS test suggests that our model's distribution is significantly different from the experimental distributions from the extremely low p-values. Therefore our results are not consistent with the cells are not arising through a Luria-Delbruck related process.
